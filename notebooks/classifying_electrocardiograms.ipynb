## Logistic Regression ðŸ“Š

### Why Logistic Regression? 
- It is a binary classificiation problem
- Based on heart beat data we need to build a model that can classifiy if a heart
beat is at risk. 
- Linear regression is not the best option for classification. 

#### Metrics 
- Accuracy 
- Recall
- Precision
- F1 

```python
# what is X's shape?
X.shape[0]
```

```python
#What is y's shape
y.shape
```

#### build the model

```python
# calling in two essentical libraries from sklearn
from sklearn.model_selection import cross_validate
from sklearn.linear_model import LogisticRegression

#establishing our scoring metrics in a list
scorers = ['accuracy', 'f1', 'precision', 'recall']

# Making a model 
model = LogisticRegression(max_iter= 10_000)

# storing results
cv_results = cross_validate(model,X,y, scoring=scorers)
```

```python
accuracy_mean = cv_results['test_accuracy'].mean()
f1_mean = cv_results['test_f1'].mean()
precision_mean = cv_results['test_precision'].mean()
recall_mean = cv_results['test_recall'].mean()
```

### Results 

```python
# Ration of correct prediction
correct_pred_ratio = accuracy_mean
correct_pred_ratio
```

```python
# Ablility to flag at-risk patients
flag_ratio = recall_mean
flag_ratio
```

```python
# Ability to flag correctly
correct_detection_ratio = precision_mean
correct_detection_ratio
```

```python
# Detecting as many at-risk patients as possibl without too many false alarms
aggregated_metric = f1_mean
aggregated_metric
```

### Observations
- The model was able to predict correctly in 94 cases out of 100 cases. 
- It was only able to capture 33.0% of at-risk patients. 
- Why? 

## Confusion Matirx ðŸ˜µ

```python
# Call in some essential libraries
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Split the dataset into training and test sets (using the hold out meathod)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train machine learning model on the training set
model.fit(X_train, y_train)

# Evaluate the performance of the model on the test set
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print("Test set accuracy:", accuracy)
```

```python
# now let's make a confusion matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_pred)
cm
```

### Results
- The confusion matrix should show that the model is influenced by the class imbalance: it predicts the heartbeats to be healthy most of the time. 
- Due to this behaviour, the model is often correct and has a high accuracy. 
- However, this also causes it to miss out on many at-risk heartbeats: 
- Thus, **it has bad recall...**
- This means the model we made is not good at the task we need to complete.
